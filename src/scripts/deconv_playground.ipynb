# Top-Down Deconvolution 
A top down deconvolution can be helpful to check if the calibration is done correctly. 
It also helps with figuring out if the simulated detector response is able to subtract the detector response correctly. 
The top-down deconvolution defined in td_deconv.py depends on the binding class, defined in binding_bins.py. If compiled for the 
first time for a given run, the cache files for the given bin width and energy window are created, which might take a few minutes. 
After that, compilation of the same energy window and bin width will be much faster. Down below you can see an 
exemplary use of the tplot_td_deconv function.
A top-down deconvolution for a given detector and a given detector response matrix is calculated and plotted. 
The options for the td_deconv are: 
- detector: str --> Choose the detector you want to plot from the detector list given in setup.py
- parity: int --> A detector response matrix depends not only on the detector and the geometry, but also on the cascade of the decay. The cascades available are $1^\mathrm{\pi} \rightarrow J^+ $. With this parameter you can set parity $\pi$ with 0 for positive and 1 for negative parity.
- state_idx: int --> Choose the final state of the cascade $J^+$. Index 0 refers to the ground state. Index 1 for the first excited state etc. Spins are adjusted automatically. 
- color_meas: str = 'black' --> Color of the measured spectrum 
color_inc: str = 'blue' --> Color of the deconvolved spectrum 
- plot_shape: tuple = (15, 10) --> Size of the plot



```python
from src.scripts.td_deconv import td_deconv

td_inst = td_deconv('Example', 179, 'bins')
td_inst.plot_td_deconv('L3', 1, 0)
```

# BinDInG deconvolution

After executing the exe_binding.py script, the results can be viewed using the plot_binding class defined in plot_binding_results.py script.
Simply execute the plot_incs function like shown below. The options for the plot_incs function are: 

- color_meas: str = 'black'--> Define color of measured spectra
- color_fit: str = 'purple'--> Define color of reconstruction
- collist_hor: list = ['red', 'darkred', 'salmon', 'firebrick'] --> List of colors for the decays  of $1^+$-states. The length of the list needs to be at least as long as how many states are given in the setup.py config. The color scheme for the $1^+$-states is traditionally chosen to be red. 
- collist_vert: list = ['blue', 'darkblue', 'cyan', 'teal'] --> Same as for the collist_hor, but for the decay of $1^-$-states. 
- color_ato: str = 'gold' --> Color of atomic scattering background 
- color_ng: str = 'green' --> Color of (n,g)-line and 16O-contamination
- beam_prof: bool = True --> Boolean to show the beam profile from the 0-degree detector deconvolution without uncertainty band. 
- beam_prof_with_unc: bool = False --> Same as beam_prof, but with uncertainty band for the beam profile. 
- beam_deconv: bool = False --> Plot the deconvolution of the 0-degree detector, including the measured spectrum and reconstruction.
- ato: bool = True --> Boolean if the atomic scattering background should be shown in the plot 
- fonts_size: int = 25 --> Set font size of the plot 
- plot_shape: tuple = (15, 20) --> Set the total plot size
- plot_arr : tuple = None --> Set the shape of the plots. The default is all detectors aligned vertically, i.e. (len(self.setup['detectors']), 1). Needs to fit to the number of detectors given in the 'dets' option. 
- dets: list = None --> List of all detectors shown in the plot. Default is all detectors given in setup.py, i.e. self.setup['detectors']. 




```python
from src.scripts.plot_binding_results import Plot_binding 

plot_inst = Plot_binding('Example', 179, 'bins')
plot_inst.plot_incs()
```
